```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib.colors import ListedColormap
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression as lr

from sklearn.pipeline import make_pipeline



#2
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/healthcare-dataset-stroke-data.csv', header= 0)

dataset.drop("Residence_type",axis=1,inplace=True)
dataset.drop("id",axis=1,inplace=True)
print(dataset.head(n=300))
```

```python

var_list = ['ever_married']
# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function
dataset[var_list] = dataset[var_list].apply(binary_map)
dataset
```

```python
var_list = ['gender']
# Defining the map function
def binary_map(x):
    return x.map({'Female': 1, "Male": 0})
# Applying the function
dataset[var_list] = dataset[var_list].apply(binary_map)
dataset
```

```python
def alter(col):
    if col == 'Private':
        return 0
    elif col == 'Self-employed':
        return 1
    elif col == 'Govt_job':
        return 2
    elif col == 'children':
        return 3
    else:
        return 4

dataset['work_type'] = dataset['work_type'].apply(alter)
dataset
```

```python
dataset['smoking_status'].unique()
def change(col):
    if col == 'formerly smoked':
        return 0
    elif col == 'never smoked':
        return 1
    elif col == 'smokes':
        return 2
    else:
        return 3

dataset['smoking_status'] = dataset['smoking_status'].apply(change)
dataset
```

# **Training and Testing**


```python
# X = dataset.iloc[:, :-1].values
# y = dataset.iloc[:, -1].values
```

```python
x1 = dataset[['hypertension', 'smoking_status', 'heart_disease', 'bmi', 'avg_glucose_level']]
y = dataset['stroke']
```

```python
#3 Splitting the dataset into the Training set and Test set
import statsmodels.api as sm
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
x1_const = sm.add_constant(x1)
x1_train, x1_test, y_train, y_test = train_test_split(x1_const, y, test_size=0.2, random_state=365)
reg = sm.Logit(y_train, x1_train)
result = reg.fit()
result.summary()
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x1_train, y_train)
```

```python
y_pred = classifier.predict(x1_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc1 = accuracy_score(y_test, y_pred)
print(f"Accuracy score: {acc1}")
```

```python
y_pred = classifier.predict(x1_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
pipe = make_pipeline(StandardScaler(), lr())
pipe.fit(x1_train,y_train)
y_pred = pipe.predict(x1_test)
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(pipe,x1_test,y_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier#loss
from sklearn import datasets
from sklearn.model_selection import learning_curve
train_sizes, train_scores, test_scores = learning_curve(classifier, x1, y, cv=10, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.01, 1.0, 50))
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.subplots(1, figsize=(10,10))
plt.plot(train_sizes, train_mean, '--', color="#111111",  label="Training score")
plt.plot(train_sizes, test_mean, color="#111111", label="Cross-validation score")

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="#DDDDDD")
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="#DDDDDD")

plt.title("Learning Curve")
plt.xlabel("Training Set Size"), plt.ylabel("Accuracy Score"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
