```python

var_list = ['ever_married']
# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function
dataset[var_list] = dataset[var_list].apply(binary_map)
dataset
```

```python
var_list = ['gender']
# Defining the map function
def binary_map(x):
    return x.map({'Female': 1, "Male": 0})
# Applying the function
dataset[var_list] = dataset[var_list].apply(binary_map)
dataset
```

```python
def alter(col):
    if col == 'Private':
        return 0
    elif col == 'Self-employed':
        return 1
    elif col == 'Govt_job':
        return 2
    elif col == 'children':
        return 3
    else:
        return 4

dataset['work_type'] = dataset['work_type'].apply(alter)
dataset
```

```python
dataset['smoking_status'].unique()
def change(col):
    if col == 'formerly smoked':
        return 0
    elif col == 'never smoked':
        return 1
    elif col == 'smokes':
        return 2
    else:
        return 3

dataset['smoking_status'] = dataset['smoking_status'].apply(change)
dataset
```

# **Training and Testing**


```python
# X = dataset.iloc[:, :-1].values
# y = dataset.iloc[:, -1].values
```

```python
x1 = dataset[['hypertension', 'smoking_status', 'heart_disease', 'bmi', 'avg_glucose_level']]
y = dataset['stroke']
```

```python
#3 Splitting the dataset into the Training set and Test set
import statsmodels.api as sm
x1_const = sm.add_constant(x1)

x1_train, x1_test, y_train, y_test = train_test_split(x1_const, y, test_size=0.2, random_state=365)
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
reg = sm.Logit(y_train, x1_train)
result = reg.fit()

result.summary()
```

```python
table_predict = pd.DataFrame()
table_predict['actual'] = y_test
table_predict['predict'] = np.round(result.predict(x1_test))
table_predict
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
cm = confusion_matrix(table_predict['actual'], table_predict['predict'])
cm
```

```python
accuracy = accuracy_score(table_predict['actual'], table_predict['predict'])
accuracy
```

```python
print('Accuracy The Model :', np.round(accuracy, 2) * 100, '%')
```
